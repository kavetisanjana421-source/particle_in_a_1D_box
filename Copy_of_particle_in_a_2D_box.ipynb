PARTICLE IN A 2D BOX:

consider a particle in a 2D box of dimensions a*b, where 0<x<a and 0<y<b inside the box the electron acts as free particle
Time independent schrodinger eq for two dimensional:
    
$$
-\frac{h^2}{2m} \left(\frac{d^2\psi(x,y)}{dx^2}+\frac{d^2\psi(x,y)}{dy^2}\right) + V(x,y)\psi(x,y)=E\psi(x,y)
$$

Boundary conditions:

--->V(x,y)=0 for 0<=x<=a and for 0<=y<=b
    
--->V(x,y)=INFINITY fot x>a,x<0 and y<0,y>b
    
so inside the box V(x,y) becomes zero and the equation becomes:

$$
-\frac{h^2}{2m} \left(\frac{d^2\psi(x,y)}{dx^2}+\frac{d^2\psi(x,y)}{dy^2}\right)=E\psi(x,y)    
$$

this equation can be expressed interms of product of two independent 1D components :
$$
\psi(x,y)=X(x).Y(y)
$$

where,
                 
X(x)=A_xsin(kx)+B_xcos(kx)
                      
Y(y)=A_ysin(ky)+B_ycos(ky)

    
Applying boundary conditions:
    
$$
\psi(0,y)=\psi(x,0)=0               
$$                      
    
$$
\psi(x,y)=N\sin(\sqrt{\frac{2mE_x}{\hbar^2}}x).\sin(\sqrt{\frac{2mE_y}{\hbar^2}}y)
$$                      

the total energy is  E=E(x)+E(y) where
                      
$$
                      \frac{\hbar^2 \pi^2}{2mL^2}*n_x^2=E(x),
                      \quad
                      \frac{\hbar^2 \pi^2}{2mL^2}*n_y^2=E(y)
$$                      
                      
hence the total energy is given by:

$$
                      \frac{\hbar^2 \pi^2}{2mL^2}*(n_x^2+n_y^2)=E(n_x,n_y)  
$$

where n_x and n_y are the two integers each starting from 1,2,3..., these arise from the independent boundary conditions in x and y directions

                      ---> therefore the ground state energy is E_11

$$                      
 \frac{\hbar^2 \pi^2}{mL^2}=E_(1,1)
$$                     

the normalized 2D wave function is

$$
    \psi_(n_x,n_y) (x,y)= \frac{2}{L} \sin(\frac{n_x \pi x}{L}) \sin(\frac{n_y \pi y}{L})
$$    
    
DEGENERACY:
        
two distinct wavefunctions are said to be degenerate if they correspond to the same energy, the most degenerate case is E_(m,n) =E_(n,m)
    

```python
import matplotlib.pyplot as plt
E = [] # Initialize E as an empty list to store calculated energy values
n = range(1, 6)

for nx in n:
  row = []
  for ny in n:
    # Calculate the energy level for each (nx, ny) pair
    energy = nx**2 + ny**2
    row.append(energy)
  E.append(row)

# Plotting the energy levels
for i in range(len(n)):
  # Access the energy values for a fixed nx (represented by i) across different ny values
  plt.plot(n, E[i], 'o-', label=(f'nx={i+1}'))

plt.title('Energy Levels of a particle in 2D box')
plt.xlabel('n_y') # Label the x-axis as n_y since we are plotting against n_y for each nx
plt.ylabel('ENERGY (in units of h^2*pi^2 / 2mL^2)') # Indicate the units of energy
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
#energy levels
energies=[] # initializing an empty list to store calculated energy values
states=[]
n=range(1,6)
for nx in n:
  for ny in n:
    E=nx**2+ny**2
    energies.append(E)
    states.append(f'({nx},{ny})')
#plotting degeneracy by highlighting same energies
plt.scatter(energies,states,color='blue')
plt.title('Degeneracy in 2D box')
plt.xlabel('Energy')
plt.ylabel('states (nx,ny)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, np.pi, 100)
y = np.linspace(0, np.pi, 100)
X, Y = np.meshgrid(x, y)

nx, ny = 1, 1
psi = np.sin(nx*X)*np.sin(ny*Y)

plt.contourf(X, Y, psi, 30, cmap='jet')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'nx={nx}, ny={ny}')
plt.show()
```
